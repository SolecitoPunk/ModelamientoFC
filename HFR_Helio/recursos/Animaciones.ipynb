```python
%%manim -v WARNING -qh Helio3B1B

class Helio3B1B(Scene):
    def construct(self):
        # --- Configuración de Estilo 3B1B ---
        COLOR_PROTON = "#E07A5F"  
        COLOR_NEUTRON = "#818589" 
        COLOR_ELECTRON = "#58C4DD" 
        COLOR_ORBITA = WHITE

        # 1. Creación del Núcleo (2p, 2n)
        p1 = Dot(color=COLOR_PROTON).shift(UP*0.1 + RIGHT*0.1)
        p2 = Dot(color=COLOR_PROTON).shift(DOWN*0.1 + LEFT*0.1)
        n1 = Dot(color=COLOR_NEUTRON).shift(UP*0.1 + LEFT*0.1)
        n2 = Dot(color=COLOR_NEUTRON).shift(DOWN*0.1 + RIGHT*0.1)
        nucleo = VGroup(p1, p2, n1, n2)
        
        label_nucleo = MathTex("2p^+, 2n^0", font_size=24).next_to(nucleo, UP)

        # 2. Órbita y Etiquetas
        orbita = Circle(radius=2, color=COLOR_ORBITA).set_stroke(opacity=0.3)
        
        titulo = Text("Átomo de Helio").scale(1)
        subtitulo = MathTex("Estado \ Base: 1s^2", color=YELLOW).next_to(titulo, DOWN)
        header = VGroup(titulo, subtitulo).to_edge(UP)

        # 3. Electrones
        e1 = Dot(color=COLOR_ELECTRON).scale(1.2)
        e2 = Dot(color=COLOR_ELECTRON).scale(1.2)

        # Posicionamos los electrones en puntos opuestos
        e1.move_to(orbita.point_from_proportion(0))
        e2.move_to(orbita.point_from_proportion(0.5))

        # --- Animación ---
        self.play(Write(header))
        self.play(FadeIn(nucleo), Write(label_nucleo))
        self.play(Create(orbita))
        self.play(FadeIn(e1), FadeIn(e2))
        
        # Rotación continua usando un Updater
        # El valor de la rotación depende del tiempo de la escena
        e1.add_updater(lambda m, dt: m.move_to(
            orbita.point_from_proportion((self.time * 0.2) % 1)
        ))
        e2.add_updater(lambda m, dt: m.move_to(
            orbita.point_from_proportion((self.time * 0.2 + 0.5) % 1)
        ))

        self.wait(5) # El video durará 5 segundos de rotación
```

```python
%%manim -v WARNING -qh AproximacionYVariacional3B1B

class AproximacionYVariacional3B1B(Scene):
    def construct(self):
        # --- CONFIGURACIÓN DE ESTILO 3B1B ---
        C_AZUL = "#58C4DD"  
        C_AMARILLO = "#FFFF00" 
        C_ROJO = "#E07A5F"   
        C_CREMA = "#FDFDD0"  

        # =========================================================
        # PARTE 1: Aproximación del Campo Central
        # =========================================================
        
        titulo1 = Text("1. Aproximación del Campo Central", font="Cambria Math").to_edge(UP)
        self.play(Write(titulo1))

        # --- Problema Real (Muchos cuerpos) ---
        def_complejo = Tex(
            r"Interacciones exactas $e^--e^-$ y $e^--Nucleo$", 
            color=C_ROJO, font_size=34
        ).next_to(titulo1, DOWN)
        self.play(FadeIn(def_complejo))

        nucleo_c = Dot(color=C_ROJO, radius=0.2).shift(LEFT*3)
        e1_c = Dot(color=C_AZUL).move_to(nucleo_c.get_center() + UP*1.5 + RIGHT*0.5)
        e2_c = Dot(color=C_AZUL).move_to(nucleo_c.get_center() + DOWN*1 + LEFT*0.8)
        
        # Usamos RED directamente para evitar errores de constantes no definidas
        int_n_e1 = DashedLine(nucleo_c, e1_c, color=RED)
        int_n_e2 = DashedLine(nucleo_c, e2_c, color=RED)
        int_e1_e2 = DashedLine(e1_c, e2_c, color=C_AMARILLO, stroke_width=5)
        atom_complejo = VGroup(nucleo_c, e1_c, e2_c, int_n_e1, int_n_e2, int_e1_e2)

        self.play(FadeIn(atom_complejo))
        self.play(e1_c.animate.shift(RIGHT*0.3), e2_c.animate.shift(UP*0.3), run_time=0.5, rate_func=wiggle)
        self.wait(1)

        flecha_trans = CurvedArrow(LEFT*1, RIGHT*1, angle=-PI/2, color=GRAY).shift(UP*0.5)
        self.play(Create(flecha_trans))

        # --- Campo Promedio ---
        def_aprox = Tex(
            r"Cada electrón siente un ``campo promedio''", 
            color=C_AZUL, font_size=34
        ).next_to(titulo1, DOWN)
        
        self.play(FadeOut(def_complejo), FadeIn(def_aprox))

        nucleo_a = Dot(color=C_ROJO, radius=0.2).shift(RIGHT*3)
        campo_promedio = Circle(radius=1.5, color=C_AZUL).move_to(nucleo_a).set_fill(C_AZUL, opacity=0.2)
        e_suave = Dot(color=C_AZUL).move_to(nucleo_a.get_center() + RIGHT*1.5)
        
        self.play(
            FadeOut(int_e1_e2),
            Transform(atom_complejo[0], nucleo_a),
            Transform(atom_complejo[1], e_suave),
            FadeOut(atom_complejo[2]),
            FadeOut(atom_complejo[3]), FadeOut(atom_complejo[4]),
            FadeIn(campo_promedio)
        )
        self.play(MoveAlongPath(atom_complejo[1], Circle(radius=1.5).move_to(nucleo_a)), run_time=2, rate_func=linear)
        self.wait(1)
        self.play(FadeOut(VGroup(titulo1, def_aprox, flecha_trans, atom_complejo[0], atom_complejo[1], campo_promedio)))

        # =========================================================
        # PARTE 2: El Método Variacional
        # =========================================================
        
        titulo2 = Text("2. Método Variacional", font="Cambria Math").to_edge(UP)
        self.play(Write(titulo2))

        def_variacional = Tex(
            r"Probar distintas ``soluciones posibles''\\buscando la energía más baja.", 
            color=C_CREMA, font_size=32
        ).next_to(titulo2, DOWN)
        self.play(FadeIn(def_variacional))

        ax = Axes(
            x_range=[0, 10, 1], y_range=[0, 10, 1],
            x_length=6, y_length=4,
            axis_config={"color": GRAY}
        ).shift(DOWN*1 + LEFT*2)
        
        curva_energia = ax.plot(lambda x: 0.2*(x-5)**2 + 2, x_range=[1, 9], color=C_AZUL)
        linea_exacta = ax.plot(lambda x: 1, x_range=[0, 10], color=C_ROJO)
        label_exacta = Tex(r"$E_{exacta}$", color=C_ROJO, font_size=24).next_to(linea_exacta, RIGHT)

        self.play(Create(ax), Create(linea_exacta), Write(label_exacta))
        self.play(Create(curva_energia))

        # --- CORRECCIÓN AQUÍ: always_redraw en minúsculas ---
        alpha_tracker = ValueTracker(2)
        
        punto_e = always_redraw(lambda: Dot(color=C_AMARILLO).move_to(
            ax.c2p(alpha_tracker.get_value(), curva_energia.underlying_function(alpha_tracker.get_value()))
        ))
        
        label_e_actual = always_redraw(lambda: MathTex(
            f"E = {curva_energia.underlying_function(alpha_tracker.get_value()):.2f}", 
            color=C_AMARILLO, font_size=30
        ).next_to(punto_e, UP))

        self.play(FadeIn(punto_e), FadeIn(label_e_actual))
        
        # Animación de búsqueda del mínimo
        self.play(alpha_tracker.animate.set_value(5), run_time=3, rate_func=smooth)
        
        flecha_min = Arrow(start=DOWN, end=UP, color=C_AMARILLO, buff=0.1).next_to(punto_e, DOWN)
        self.play(Create(flecha_min))
        self.play(Indicate(label_e_actual))
        self.wait(2)
```

```python
%%manim -v WARNING -qh AutoconsistenciaHF

class AutoconsistenciaHF(Scene):
    def construct(self):
        # --- CONFIGURACIÓN DE ESTILO 3B1B ---
        C_CYAN = "#58C4DD"  
        C_AMARILLO = "#FFFF00" 
        C_ROJO = "#E07A5F"   
        C_CREMA = "#FDFDD0"  
        C_GRIS = "#888888"

        # =========================================================
        # PARTE 1: Concepto de Ciclo de Autoconsistencia
        # =========================================================
        
        titulo = Text("Autoconsistencia: El Ciclo de Hartree-Fock", font="Cambria Math").to_edge(UP)
        self.play(Write(titulo))

        # Diagrama de bloques
        bloque_guess = Tex("Guess: $C^{(0)}$", color=C_GRIS).shift(LEFT*4 + DOWN*1)
        bloque_fock = Tex("Construir Matriz $F$", color=C_CREMA).shift(UP*0.5)
        bloque_nuevo = Tex("Nuevo: $C^{(n+1)}$", color=C_CYAN).shift(RIGHT*4 + DOWN*1)
        
        # Flechas de flujo (Usamos Create para evitar el error de GrowArrow)
        f1 = Arrow(bloque_guess.get_top(), bloque_fock.get_left(), color=C_GRIS)
        f2 = Arrow(bloque_fock.get_right(), bloque_nuevo.get_top(), color=C_CYAN)
        f_loop = CurvedArrow(bloque_nuevo.get_bottom(), bloque_guess.get_bottom(), angle=-PI/2, color=C_CYAN)

        self.play(FadeIn(bloque_guess), FadeIn(bloque_fock), FadeIn(bloque_nuevo))
        self.play(Create(f1), Create(f2))
        self.play(Create(f_loop))
        
        txt_loop = Tex("¿Ha convergido?", color=C_AMARILLO, font_size=30).next_to(f_loop, DOWN)
        self.play(Write(txt_loop))
        self.wait(1)

        # Limpieza
        self.play(FadeOut(VGroup(bloque_guess, bloque_fock, bloque_nuevo, f1, f2, f_loop, txt_loop)))

        # =========================================================
        # PARTE 2: Simulación Iterativa (Basada en el PDF)
        # =========================================================
        
        label_iter = Text("Iteración: ", font_size=30).to_edge(UL).shift(DOWN)
        num_iter = Integer(0).next_to(label_iter, RIGHT)
        self.play(Write(VGroup(label_iter, num_iter)))

        vector_label = MathTex("C^{(n)} = ", font_size=36).shift(LEFT*3 + UP*1)
        
        # --- CORRECCIÓN DE F-STRING PARA LATEX ---
        def get_vector_tex(n):
            if n == 0:
                # Usamos dobles llaves {{ }} para los entornos de LaTeX
                return r"\begin{pmatrix} 1.00 \\ 1.00 \\ 1.00 \\ 1.00 \end{pmatrix}"
            val = 1.0 + 0.5/(n+1)
            # En f-strings: {{ }} para LaTeX, { } para variables de Python
            return rf"\begin{{pmatrix}} {val:.2f} \\ {val/2:.2f} \\ {val/4:.2f} \\ {val/8:.2f} \end{{pmatrix}}"

        vector_val = MathTex(get_vector_tex(0), color=C_CYAN).next_to(vector_label, RIGHT)
        self.play(Write(vector_label), FadeIn(vector_val))

        # Átomo visual
        nucleo = Dot(color=C_ROJO)
        nube = always_redraw(lambda: 
            Circle(radius=1.5, color=C_CYAN).set_fill(C_CYAN, opacity=0.1 + (num_iter.get_value()*0.1))
        )
        self.play(FadeIn(nucleo), Create(nube))

        # Ciclo de iteraciones
        for i in range(1, 5):
            self.play(num_iter.animate.set_value(i), run_time=0.4)
            
            fock_eq = MathTex(r"F(C^{(n)}) C^{(n+1)} = \epsilon S C^{(n+1)}", color=C_AMARILLO).scale(0.7).shift(DOWN*2)
            self.play(FadeIn(fock_eq), Flash(fock_eq, color=C_AMARILLO))
            
            new_vector = MathTex(get_vector_tex(i), color=C_CYAN).next_to(vector_label, RIGHT)
            self.play(Transform(vector_val, new_vector), FadeOut(fock_eq))
            self.wait(0.2)

        # Final
        check = Tex("¡Solución Autoconsistente!", color=C_AMARILLO).shift(DOWN*2.5)
        self.play(Write(check), Indicate(vector_val))
        self.wait(3)
```

```python
%%manim -qh -v WARNING FermiNetExplanation

from manim import *

class FermiNetExplanation(Scene):
    def construct(self):
        # Título
        title = Text("FermiNet", font_size=72, color=BLUE)
        subtitle = Text("Redes neuronales para la ecuación de Schrödinger electrónica",
                        font_size=36, color=WHITE)
        subtitle.next_to(title, DOWN)
        self.play(Write(title), Write(subtitle))
        self.wait(2)
        self.play(FadeOut(title), FadeOut(subtitle))

        # 1. Átomo de Helio con dos electrones (modelo clásico)
        nucleus = Circle(radius=0.2, color=RED, fill_opacity=1).shift(UP * 0.5)
        nucleus_label = Text("Núcleo", font_size=24).next_to(nucleus, UP)
        electron1 = Dot(color=YELLOW).move_to(nucleus.get_center() + RIGHT * 1.5 + UP * 0.5)
        electron2 = Dot(color=YELLOW).move_to(nucleus.get_center() + LEFT * 1.2 + DOWN * 0.3)
        electron1_label = Text("e⁻", font_size=24).next_to(electron1, UR)
        electron2_label = Text("e⁻", font_size=24).next_to(electron2, DL)

        self.play(FadeIn(nucleus), FadeIn(nucleus_label),
                  FadeIn(electron1), FadeIn(electron2),
                  FadeIn(electron1_label), FadeIn(electron2_label))
        self.wait(1)

        # 2. Hartree-Fock: campo medio → dos orbitales independientes
        mean_field_text = Text("Hartree-Fock: Campo medio", font_size=36, color=GREEN)
        mean_field_text.to_corner(UL)
        self.play(Write(mean_field_text))

        # Representación de orbitales como nubes difusas (círculos semitransparentes)
        orbital1 = Circle(radius=1.2, color=BLUE, fill_opacity=0.2, stroke_width=2).move_to(electron1)
        orbital2 = Circle(radius=1.2, color=BLUE, fill_opacity=0.2, stroke_width=2).move_to(electron2)
        self.play(Create(orbital1), Create(orbital2))
        self.wait(2)

        # Mostrar que la probabilidad es producto de dos densidades independientes
        prob_indep = MathTex(r"P(\mathbf{r}_1, \mathbf{r}_2) = |\phi_{1s}(\mathbf{r}_1)|^2 |\phi_{1s}(\mathbf{r}_2)|^2",
                             font_size=36)
        prob_indep.next_to(mean_field_text, DOWN, buff=0.5)
        self.play(Write(prob_indep))
        self.wait(2)

        # 3. Limitación: correlación ignorada (los electrones pueden estar cerca)
        arrow = Arrow(orbital1.get_bottom(), orbital2.get_top(), buff=0.2, color=RED)
        problem_text = Text("No hay correlación: ¡pueden estar muy cerca!", font_size=30, color=RED)
        problem_text.next_to(arrow, RIGHT)
        self.play(GrowArrow(arrow), Write(problem_text))
        self.wait(2)

        # Mover electrones cerca para mostrar el problema
        self.play(electron1.animate.move_to(nucleus.get_center() + RIGHT * 0.8),
                  electron2.animate.move_to(nucleus.get_center() + LEFT * 0.6),
                  orbital1.animate.move_to(electron1),
                  orbital2.animate.move_to(electron2),
                  run_time=2)
        self.wait(1)

        # 4. Introducción a FermiNet
        self.play(FadeOut(arrow), FadeOut(problem_text), FadeOut(orbital1), FadeOut(orbital2))
        ferminet_title = Text("FermiNet", font_size=48, color=BLUE)
        ferminet_title.move_to(mean_field_text.get_center())
        self.play(Transform(mean_field_text, ferminet_title))
        self.wait(1)

        # Diagrama de la red neuronal
        # Capa de entrada: coordenadas de los electrones
        input_neurons = VGroup(*[Dot(radius=0.1, color=WHITE) for _ in range(6)])
        input_neurons.arrange(DOWN, buff=0.4).next_to(nucleus, RIGHT * 3)
        input_label = Text("Entrada:\ncoordenadas", font_size=24).next_to(input_neurons, LEFT)

        # Capas ocultas (representadas como bloques)
        hidden_layer1 = Rectangle(height=2.5, width=1.2, color=YELLOW, fill_opacity=0.5).next_to(input_neurons, RIGHT, buff=1.5)
        hidden_label1 = Text("Capa oculta 1", font_size=20).next_to(hidden_layer1, UP)
        hidden_layer2 = Rectangle(height=2.5, width=1.2, color=YELLOW, fill_opacity=0.5).next_to(hidden_layer1, RIGHT, buff=1.5)
        hidden_label2 = Text("Capa oculta 2", font_size=20).next_to(hidden_layer2, UP)

        # Capa de salida: el valor de la función de onda
        output = Rectangle(height=1.2, width=1.2, color=GREEN, fill_opacity=0.5).next_to(hidden_layer2, RIGHT, buff=1.5)
        output_label = Text("Ψ( r₁, r₂ )", font_size=24).next_to(output, UP)

        # Conectar con flechas (solo algunas para no saturar)
        arrow_in_h1 = Arrow(input_neurons.get_right(), hidden_layer1.get_left(), buff=0.2)
        arrow_h1_h2 = Arrow(hidden_layer1.get_right(), hidden_layer2.get_left(), buff=0.2)
        arrow_h2_out = Arrow(hidden_layer2.get_right(), output.get_left(), buff=0.2)

        self.play(FadeIn(input_neurons), FadeIn(input_label),
                  FadeIn(hidden_layer1), FadeIn(hidden_label1),
                  FadeIn(hidden_layer2), FadeIn(hidden_label2),
                  FadeIn(output), FadeIn(output_label),
                  Create(arrow_in_h1), Create(arrow_h1_h2), Create(arrow_h2_out))
        self.wait(2)

        # 5. Flujo de información: cada electrón ve a los demás
        info_flow_text = Text("Cada electrón recibe información de los demás",
                              font_size=30, color=YELLOW)
        info_flow_text.next_to(hidden_layer2, UP, buff=1)
        self.play(Write(info_flow_text))

        # Dibujar dos electrones y líneas de interacción
        e1_dot = Dot(color=YELLOW).move_to(ORIGIN + LEFT * 2 + DOWN * 1)
        e2_dot = Dot(color=YELLOW).move_to(ORIGIN + RIGHT * 2 + DOWN * 1)
        e1_label = Text("e⁻₁", font_size=20).next_to(e1_dot, DOWN)
        e2_label = Text("e⁻₂", font_size=20).next_to(e2_dot, DOWN)
        self.play(FadeIn(e1_dot), FadeIn(e2_dot), FadeIn(e1_label), FadeIn(e2_label))

        # Flechas que conectan cada electrón con la red
        arrow_e1_net = Arrow(e1_dot.get_top(), input_neurons.get_bottom(), buff=0.2, color=BLUE)
        arrow_e2_net = Arrow(e2_dot.get_top(), input_neurons.get_bottom(), buff=0.2, color=BLUE)
        self.play(Create(arrow_e1_net), Create(arrow_e2_net))
        self.wait(2)

        # 6. Correlación: los electrones se repelen → "agujero de correlación"
        correlation_text = Text("Correlación: los electrones se evitan mutuamente",
                                font_size=30, color=GREEN)
        correlation_text.move_to(info_flow_text)
        self.play(Transform(info_flow_text, correlation_text))

        # Mostrar dos electrones que se repelen: animación de separación
        self.play(e1_dot.animate.shift(LEFT * 1.5), e2_dot.animate.shift(RIGHT * 1.5), run_time=2)
        self.wait(1)

        # 7. Resultado: función de onda aprendida
        final_text = Text("FermiNet aprende la función de onda\nque minimiza la energía",
                          font_size=36, color=WHITE)
        final_text.move_to(ORIGIN)
        self.play(FadeOut(arrow_e1_net), FadeOut(arrow_e2_net), FadeOut(e1_dot), FadeOut(e2_dot),
                  FadeOut(e1_label), FadeOut(e2_label),
                  FadeOut(input_neurons), FadeOut(input_label),
                  FadeOut(hidden_layer1), FadeOut(hidden_label1),
                  FadeOut(hidden_layer2), FadeOut(hidden_label2),
                  FadeOut(output), FadeOut(output_label),
                  FadeOut(arrow_in_h1), FadeOut(arrow_h1_h2), FadeOut(arrow_h2_out),
                  FadeOut(nucleus), FadeOut(nucleus_label), FadeOut(electron1), FadeOut(electron2),
                  FadeOut(electron1_label), FadeOut(electron2_label), FadeOut(prob_indep))
        self.play(Write(final_text))
        self.wait(3)
        self.play(FadeOut(final_text))
```
